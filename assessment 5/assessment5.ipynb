# Part 1: Working with variables, input and strings

### 1: Prompt the user to enter their full name and store it in a variable. Then, print the greeting message.

```python
name = input("Enter your full name: ")
print("Hello", name, "! Welcome to the Python assignment.")
```

### 2: String method

```python
print("Your name in upper case is: ", name.upper())
print("Your name in lower case is: ", name.lower())
print("Length of your name is: ", len(name))
print("Name with hyphen is: ", name.replace(" ", "-"))
```

### 3: Calculate the userâ€™s approximate age by subtracting the birth year from the current year

```python
print("Enter your birth year: ")
birth_year = int(input())
current_year = 2026
age = current_year - birth_year
print("Current year is",current_year,"Your approximate age is: ", age)
```

# Part 2: List and Indexing

### 1: create a list from user input

```python
fruit_input = input("Enter three of your favorite fruits (comma-separated): ")
fruit_list = fruit_input.split(',')
print("Your list of fruits:", fruit_list)
```

### 2: List indexing and slicing

```python
print("First fruit:", fruit_list[0])
print("Last fruit:", fruit_list[-1])
print("Second fruit:", fruit_list[1])
print("First two fruits:", fruit_list[:2])
```

### 3: List methods

```python
print("Adding a new fruit...")
fruit_list.append("strawberry")
print("Updated list:", fruit_list)
print("Removing one fruit...")
fruit_list.pop(2) 
print("Updated list:", fruit_list)
print("Length of the list is now", len(fruit_list))
```
