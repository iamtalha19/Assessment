## Question 2: Swap Two Variables (Using Third Variable)

```python
print("Enter first number: ")
num1 = int(input())
print("Enter second number: ")
num2 = int(input())

print("Before swapping\n first number: ", num1, "\n second number: ", num2)
num3 = num1
num1 = num2
num2 = num3
print("After swapping\n first number: ", num1, "\n second number: ", num2) 
```

## Question 3: Swap Two Variables (Without Using Third Variable)

```python
print("Enter first number: ")
num1 = int(input())
print("Enter second number: ")
num2 = int(input())

print("Before swapping\n first number: ", num1, "\n second number: ", num2)
num1 = num1 + num2
num2 = num1 - num2
num1 = num1 - num2
print("After swapping\n first number: ", num1, "\n second number: ", num2)
```

## Question 4: Convert Temperature from Celsius to Fahrenheit

```python
print("Enter temperature in Celsius: ")
celsius = float(input())
fahrenheit = (celsius * 9/5) + 32
print("Temperature in Fahrenheit: ", fahrenheit)
```
