# Python Functions Practice

#### 1. Student Grade Calculator

```python
def calculate_grade(marks):
    if marks >= 90:
        return "A"
    elif marks >= 80:
        return "B"
    elif marks >= 70:
        return "C"
    else:
        return "Fail"

print(calculate_grade(80))
```

#### 2. Even or Odd Number Checker

```python
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
print(is_even(4))
```

#### 3. Factorial Using Function

```python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result
print(factorial(5))
```

#### 4. Find Maximum Number

```python
def find_max(a, b, c):
    if a > b and a > c:
        return a
    elif b > a and b > c:
        return b
    else:
        return c
print(find_max(1, 2, 3))
```

#### 5. Word Counter

```python
def count_words(sentence):
    words = sentence.split()
    return len(words)

print(count_words("This is a sample sentence"))
```

#### 6. Prime Number Checker

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
print(is_prime(5))
```

#### 7. List Sum Function

```python
def list_sum(numbers):
    total = 0
    for number in numbers:
        total += number
    return total
print(list_sum([10, 20, 30]))
```

#### 8. Password Strength Checker

```python
def check_password(password):
    if len(password) < 8:
        return "Weak"
    for char in password:
        if char.isdigit():
            return "Strong"
    return "Weak"
print(check_password("password"))
```

#### 9. Temperature Converter

```python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32
print(celsius_to_fahrenheit(0))
```

#### 10. Remove Duplicates from List

```python
def remove_duplicates(data):
    return list(set(data))
print(remove_duplicates([1, 2, 2, 3, 4, 4]))
```
