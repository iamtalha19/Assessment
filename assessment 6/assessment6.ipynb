### Part 1: Money Left After Purchase

```python
initial_money = 50
item_price = 15
tax_rate = 0.03
total_cost = item_price * (1 + tax_rate)
money_left = initial_money - total_cost
print(f"Money left after purchase: ${money_left}")
```

### Part 2: Approximate Weeks Until Birthday

```python
days_until_birthday = int(input("How many days until your birthday? "))
weeks_until_birthday = days_until_birthday / 7
print(f"Approximate weeks until your birthday: {weeks_until_birthday}")
```

### Part 3: Letter Grade Conversion

```python
Grade = int(input("Enter your grade: "))
if 90 <= Grade <= 100:
    letter = 'A'
elif 80 <= Grade <= 89:
    letter = 'B'
elif 70 <= Grade <= 79:
    letter = 'C'
elif 60 <= Grade <= 69:
    letter = 'D'
else:
    letter = 'F'
print(f"Numerical Grade: {Grade}")
print(f"Letter Grade: {letter}")
```
