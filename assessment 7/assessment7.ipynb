# Python For Loop Questions

#### 1. Write a program to print 1 to 50 using for loop

````python
for i in range(1,51):
    print(i)
````

#### 2. Print all even numbers between 1 and 100 using for loop

````python
for i in range(1,101):
    if i%2==0:
        print(i)
````

#### 3. Write a program to calculate the sum of number from 1 to N entered by user

````python
N=int(input("Enter a number: "))
sum=0
for i in range(1,N+1):
    sum+=i
print("The sum of number from 1 to",N,"is",sum)
````

#### 4. Print the multiplication table of a number entered by user (upto 10)

````python
num=int(input("Enter a number: "))
for i in range(1,11):
    print(num,"*",i,"=",num*i)
````

#### 5. Write a program to reverse a string using for loop

````python
string=input("Enter a string: ")
reverse=""
for i in string:
    reverse=i+reverse
print("The reverse of '",string,"' is '",reverse,"' ")
````

#### 6. Count the number of vowels present in a string using for loop

````python
string=input("Enter a string: ")
vowels="aeiou"
count=0
for i in string:
    if i in vowels:
        count+=1
print("The number of vowels in the string is",count)
````

#### 8. Write a program to check whether a number is prime using a for loop

````python
num=int(input("Enter a number: "))
for i in range(2,num):
    if num%i==0:
        print(num,"is not a prime number")
        break
else:
    print(num,"is a prime number")
````

#### 9. Given a list: `numbers = [12, 5, 7, 18, 21, 30]`
**Print only numbers divisible by 3.**

````python
numbers = [12, 5, 7, 18, 21, 30]
for i in numbers:
    if i%3==0:
        print(i)
````

#### 10. Break and Continue
**Print numbers from 1 to 20 but: Skip multiples of 3. Stop execution if number becomes greater than 15.**

````python
for i in range(1,21):
    if i%3==0:
        continue
    if i>15:
        break
    print(i)
````

#### 11. Calculate factorial of a number using a while loop

````python
num=int(input("Enter a number: "))
factorial=1
original_num = num
while num>0:
    factorial*=num
    num-=1
print("The factorial of",original_num,"is",factorial)
````
